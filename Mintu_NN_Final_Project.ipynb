```python
# Import packages

import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
# Drop the columns 

group3_hype = group3_hype.drop(['INTAKE TERM CODE', 'ADMIT TERM CODE', 'ID 2','RECORD COUNT','PROGRAM LONG NAME','STUDENT LEVEL NAME',
                                'STUDENT TYPE NAME', 'STUDENT TYPE GROUP NAME',
       'TIME STATUS NAME', 'RESIDENCY STATUS NAME', 'FUNDING SOURCE NAME',
       'MAILING CITY NAME','MAILING POSTAL CODE', 'MAILING PROVINCE NAME','MAILING COUNTRY NAME','FUTURE TERM ENROL','EXPECTED GRAD TERM CODE',
       'HS AVERAGE GRADE','HS AVERAGE MARKS','FIRST GENERATION IND', 'APPL FIRST LANGUAGE DESC',
                        'APPL EDUC INST TYPE NAME', 'APPLICANT TARGET SEGMENT NAME', 'PREV EDU CRED LEVEL NAME'],axis=1)
```

```python
# Shape of the dataset after dropping the columns

group3_hype.shape
```

```python
#dropping the rows with In Progress values

group3_hype = group3_hype[group3_hype['SUCCESS LEVEL'].str.contains("In Progress")==False]
```

```python
#target column after dropping the rows

group3_hype['SUCCESS LEVEL'].value_counts()
```

```python
# Data information

group3_hype.info()
```

```python
# Dropping records with null value

group3_hype = group3_hype.dropna()
```

```python
# Shape of the dataset after dropping the records with null values

group3_hype.shape
```

```python
#counting null values

group3_hype.isnull().sum()
```

```python
#categorical data

categorical_cols = ['INTAKE COLLEGE EXPERIENCE','PRIMARY PROGRAM CODE','SCHOOL CODE','MAILING POSTAL CODE GROUP 3','GENDER','DISABILITY IND','CURRENT STAY STATUS',
                    'APPLICANT CATEGORY NAME'] 

group3_hype = pd.get_dummies(group3_hype, columns = categorical_cols)
```

```python
# print dataset

group3_hype.head(10)
```

```python
# Converted all rest columns using Label encoder.

group3_hype['AGE GROUP LONG NAME'] = group3_hype['AGE GROUP LONG NAME'].apply(lambda x: ['0 to 18', '19 to 20','21 to 25', '26 to 30','31 to 35', '36 to 40', '41 to 50'].index(x))
group3_hype['SUCCESS LEVEL'] = group3_hype['SUCCESS LEVEL'].apply(lambda x: ['Unsuccessful', 'Successful'].index(x))
group3_hype['ACADEMIC PERFORMANCE'] = group3_hype['ACADEMIC PERFORMANCE'].apply(lambda x: ['ZZ - Unknown','DF - Poor','C - Satisfactory','AB - Good'].index(x))
```

```python
# print dataset

group3_hype.head(5)
```

```python
# Normalizing the all the data.

x = group3_hype.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
group3_hype_norm = pd.DataFrame(x_scaled)
group3_hype_norm.columns = group3_hype.columns
```

```python
group3_hype_norm.head(5)
```

```python


y = group3_hype['SUCCESS LEVEL']
X = group3_hype.drop(['SUCCESS LEVEL'],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=75)
```

```python

def get_basic_model():
  
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='sigmoid',input_dim=X_train.shape[1]),
    tf.keras.layers.Dense(5, activation='relu'),
    tf.keras.layers.Dense(1)
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=['accuracy'])
  return model
```

```python
model = get_basic_model()
BATCH_SIZE = 2

history = model.fit(tf.convert_to_tensor(X_train), tf.convert_to_tensor(y_train), epochs=50)
```

```python
# evaluate the keras model

_, accuracy = model.evaluate(X_train, y_train)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# Prediction

y_pred = model.predict(X_test)

score = model.evaluate(X_test, y_test,verbose=1)

print(score)
```
