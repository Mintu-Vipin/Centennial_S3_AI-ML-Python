**3. Use PCA to retrieve the 1th and 2nd principal component and output their explained variance ratio**

```python
mnist_data_X,mnist_data_y = mnist_data.iloc[:, :-1], mnist_data.iloc[:, [-1]]
```

```python
from sklearn.decomposition import PCA

pca = PCA()
mnist_data_X2D = pca.fit(mnist_data_X)
```

```python
# 1st principal component 
pca_1 = mnist_data_X2D.components_.T[:,0]
print(pca_1)

# 2nd principal component 
pca_2 = mnist_data_X2D.components_.T[:,1]
print(pca_2)
```

```python
# explained variance ratio of 1 st principal component
print(pca.explained_variance_ratio_[0])
# explained variance ratio of 1 st principal component
print(pca.explained_variance_ratio_[1])
```

**4. Plot the projections of the 1th and 2nd principal component onto a 1D hyperplane**

```python
import matplotlib.pyplot as pl
import numpy as np

# 1 st principal component
plt.figure(figsize=(10,6))
plt.scatter(pca_1, np.zeros_like(pca_1))
plt.title("First Principal Component")
plt.show()

# @ nd principal component
plt.figure(figsize=(10,6))
plt.scatter(pca_2, np.zeros_like(pca_1))
plt.title("Second Principal Component")
plt.show()
```

**5. Use Incremental PCA to reduce the dimensionality of the MNIST dataset down to 154 dimensions**

```python
from sklearn.decomposition import IncrementalPCA

n_batch = 100
inc_pca = IncrementalPCA(n_components=154)
for X_batch in np.array_split(mnist_data_X, n_batch):
    inc_pca.partial_fit(X_batch)

X_reduced = inc_pca.transform(mnist_data_X)
```

**1. Generate Swiss roll dataset**

```python
from sklearn.datasets import make_swiss_roll

n_samples, noise, random_state=1000, 0.00, 1
swiss_X, swiss_y = make_swiss_roll(n_samples, noise, random_state)
```

```python
print(swiss_X.shape, swiss_y.shape, swiss_X[:,0].shape)
print("_________________________")
print(swiss_X, swiss_y)
```

**2. Plot the resulting generated Swiss roll dataset**

```python
# Plotting the data using Matpoltlib

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Plot result
fig = plt.figure()

fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(swiss_X[:, 0], swiss_X[:, 1], swiss_X[:, 2], c = swiss_y, cmap=plt.cm.rainbow)

ax.set_title("Swiss roll in 3D")
plt.show()
```

**3. Kernel PCA (kPCA) with linear kernel, a RBF kernel, and a sigmoid kernel**

```python
from sklearn.decomposition import KernelPCA

# kPCA with linear kernel
linear_kpca = KernelPCA(n_components =2,kernel = "linear")
linearX_transformed = linear_kpca.fit_transform(swiss_X)
print(linearX_transformed.shape)

# kPCA with RBF kernel
rbf_kpca = KernelPCA(n_components=2,kernel="rbf", gamma= 0.1)
rbfX_transformed = rbf_kpca.fit_transform(swiss_X)
print(rbfX_transformed.shape)

# kPCA with sigmoid kernel
sigmoid_kpca = KernelPCA(n_components=2,kernel="sigmoid")
sigmoidX_transformed = sigmoid_kpca.fit_transform(swiss_X)
print(sigmoidX_transformed.shape)
```

**4. Plot the kPCA results of applying the linear kernel, a RBF kernel , and a sigmoid kernel from (3). Explain and compare the results **


1.   Linear : Reduce the dimensions of the dataset with minimal loss of information.

2.   Non Linear: Most real world data requires nonlinear methods in order to perform tasks that involve the analysis and discovery of patterns successfully. In the case of linearly inseparable data, a nonlinear technique is required if the task is to reduce the dimensionality of a dataset.

*   The basic idea to deal with linearly inseparable data is to project it onto a higher dimensional space where it becomes linearly separable. Let us call this nonlinear mapping function ϕ so that the mapping of a sample x can be written as x→ϕ(x), which is called “kernel function.”

In our analysis, non linear keranl PCA could project the data very well especially with RBF kernel. We can see the shape of the each graph below.














```python
# kPCA with linear kernel
plt.figure(figsize=(8,6))
plt.title("kPCA with linear kernel")
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(linearX_transformed[:,0], linearX_transformed[:,1], c = swiss_y, cmap=plt.cm.rainbow)
plt.show()

# kPCA with RBF kernel
plt.figure(figsize=(8,6))
plt.title("kPCA with RBF kernel")
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(rbfX_transformed[:,0], rbfX_transformed[:,1], c = swiss_y, cmap=plt.cm.rainbow)
plt.show()

# kPCA with sigmoid kernel
plt.figure(figsize=(8,6))
plt.title("kPCA with sigmoid kernel")
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(sigmoidX_transformed[:,0], sigmoidX_transformed[:,1], c = swiss_y, cmap=plt.cm.rainbow)
plt.show()
```
